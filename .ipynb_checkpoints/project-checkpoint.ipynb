# Sleep Pattern Analysis
## Introduction
For this project I am analyzing data gathered from my phone regarding my step count and my sleep patterns. Using an alarm clock app called Sleep Cycle I have been able to collect data about how long I sleep and how good that sleep is. It collects this data by using the phones microphone to determine how much I move around in my sleep. For the step count I'm using IOS's built in step count recorder. This data is found in the health app and can be exported to a XML file. I'm going to use linear regression to try to find correlations between step count, sleep quality, and time in bed. I'm also going to look at how the day of week is related to these metrics.

## Data Analysis
Since the Apple Health data comes in an XML file the first step was getting that converted to csv so I could load it into a dataframe. I used this site to convert Apple's XML health data to csv: https://www.ericwolter.com/projects/apple-health-export/
The Sleep Cycle data came as a csv, but it used semicolons as the delimieter so before I first replaced all the semicolons with commas and then I loaded the data into a dataframe. I dropped the columns that didn't contain any data or were irrelevant to my research.

```python
import pandas as pd
import calendar

ifile = open("sleepdata.csv", "r")
ofile = open("sleepData_clean.csv", "w")
text = ifile.read()
text = text.replace(";", ",")
ofile.write(text)
ifile.close()
ofile.close()

df = pd.read_csv("sleepData_clean.csv")
df.drop(columns=['Wake up', 'Sleep Notes', 'Heart rate', 'Activity (steps)'], inplace=True)

df
```

As you can see the resulting data frame has four columns: Start, End, Sleep quality, and Time in bed. I want to split the start and end columns so they just have time, and then make a new column that has the date. To do this I split start and end columns into year, month, and day. To determine the date I only looked at "End" because I may go to sleep before or after midnight (so this won't help us know which day the night sleep is referring to). This means the data and day of week columns refer to the day that I wake up, not the day that I go to sleep. I used these date columns to add another column that contains the day of week for each row.

```python
years = []
months = []
days = []
day_of_weeks = []
weekday_key = {0 : "Monday", 1 : "Tuesday", 2: "Wednesday", 3 : "Thursday", 4 : "Friday", 5 : "Saturday", 6 : "Sunday"}

for item in df["End"]:
    years.append(item[0:4])
    months.append(item[5:7])
    days.append(item[8:10])
    day_of_weeks.append(weekday_key[calendar.weekday(int(years[-1]), int(months[-1]), int(days[-1]))])

for i in range(len(df)):
    df["Start"][i] = df["Start"][i][11 :]
    df["End"][i] = df["End"][i][11 :]
    df["Sleep quality"][i] = float(df["Sleep quality"][i].replace("%", ""))
    
df["year"] = years
df["month"] = months
df["day"] = days
df["day of week"] = day_of_weeks

df
```

The next step is to create a new data frame using the step count data.

```python
step_df = pd.read_csv("stepcount.csv", sep=";")
step_df
```

You can see that this data frame has a ton of information that isn't relevant to me so I will drop these columns.

```python
step_df.drop(columns=["type", "sourcename", "sourceversion", "device", "unit", "creationdate"], inplace=True)
step_df
```

You can see that each column isn't associated with a specefic day, it's a 10 minute time frame. I have to combine all the rows that are of the same day to get the total step count for each day. To do this I created three new columns (year, month, day).

```python
def split_date(df, column):
    years = []
    months = []
    days = []
    day_of_weeks = []
    weekday_key = {0 : "Monday", 1 : "Tuesday", 2: "Wednesday", 3 : "Thursday", 4 : "Friday", 5 : "Saturday", 6 : "Sunday"}

    for item in df[column]:
        years.append(item[0:4])
        months.append(item[5:7])
        days.append(item[8:10])
        day_of_weeks.append(weekday_key[calendar.weekday(int(years[-1]), int(months[-1]), int(days[-1]))])
    
    df["year"] = years
    df["month"] = months
    df["day"] = days
    df["day of week"] = day_of_weeks
    
    return df
    
step_df = split_date(step_df, "startdate")
step_df
```

Now that I have the date for each row I can drop the startdate and enddate columns.

```python
step_df.drop(columns=["startdate","enddate"], inplace=True)
step_df
```

Now I will use groupby() to find the total number of steps for each day and put that into a new data frame.

```python
step_df_grouped = step_df.groupby(["year","month","day"])

years = []
months = []
days = []
steps = []
for name,item in step_df_grouped:
    years.append(name[0])
    months.append(name[1])
    days.append(name[2])
    steps.append(item.sum()["value"])
d = {"year":years, "month":months, "day":days, "steps":steps}
new_df = pd.DataFrame(data=d)
new_df
```

The next step is to merge the step count and sleep data dataframes.

```python
data = df.merge(new_df)
data
```

You can see that the "Time in bed" data is in the format hours:minutes. To be able to perform calculation on this I want to convert it to just minutes.

```python
# convert time in bed to int
for i in range(len(data)):
    data["Time in bed"][i] = 60 * int(data["Time in bed"][i][:-3]) + int(data["Time in bed"][i][-2:])
```

Next I will use linear regression to model number of steps as a function of the amount of sleep I got. However, based on my knowledge of my own sleep schedule, I think it is very likely that the time that I fall asleep and wake up will be largely determined by the day of week it is. To determine this I will have to convert the start and end data into a float.

```python
#print(data["End"][0])
start = []
end = []
for item in data["Start"]:
    start.append(float(item[:2]) + float(item[3:5]) / 60)
for item in data["End"]:
    end.append(float(item[:2]) + float(item[3:5]) / 60)
data["Start"] = start
data["End"] = end
data
```

Now I need to change figure out some what to format these columns so they numerically depict their true relations. ie 23 is only one hour away from 1, but if I perform anything on them as is it wouldn't look that way. I will represent pm times as hours before midnight, so 11pm is -1.

```python
start = []

for item in data["Start"]:
    if item > 12:
        start.append(item - 24)
    else:
        start.append(item)
        
data["Start"] = start
data
```

```python
grouped_by_day = data.groupby("day of week")
monday_df = grouped_by_day.get_group("Monday")
tuesday_df = grouped_by_day.get_group("Tuesday")
wednesday_df = grouped_by_day.get_group("Wednesday")
thursday_df = grouped_by_day.get_group("Thursday")
friday_df = grouped_by_day.get_group("Friday")
saturday_df = grouped_by_day.get_group("Saturday")
sunday_df = grouped_by_day.get_group("Sunday")

x = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
y = [saturday_df["Start"].mean(), sunday_df["Start"].mean(), monday_df["Start"].mean(), tuesday_df["Start"].mean(), wednesday_df.mean()["Start"].mean(), thursday_df["Start"].mean(), friday_df["Start"].mean()]

plt.figure()
plt.bar(x, y)
plt.title("Time Asleep by Day")
plt.ylabel("Time (hours away from midnight)")
plt.show()
```

This graph is suprising to me because it indicated that I tend to fall asleep around midnight every day, except for on Sunday nights. I know that I stay up later on Friday or Saturday nights, so this can't be right. While thinking about it I realized that on weekend nights when I go out and do stuff (and hence stay up late) I probably wouldn't even use the Sleep Cycle app at all, so the data isn't representive of every night, only nights where I don't go out. It does make sense that I stay up late on Sundays because I usually have to do an entire DA that I put off all week.
